# Equações Diferenciais EXEMPLO1

```python
import numpy as np
from scipy.integrate import odeint
from math import log
import matplotlib.pyplot as plt


# Variáveis
# N(t) -> Número de partículas radioativas
# t -> Tempo em anos

# Define a function which calculates the derivative
def dN_dt(N, t):
    TempoMeiaVida=30.0
    k=log(2)/TempoMeiaVida
    return -k*N

t = np.linspace(0,100,101)
N0 = 1.0 # the initial condition

# Função, condição inicial e variável independente
N = odeint(dN_dt, N0, t)
# N = np.array(N).flatten()

# Plot the numerical solution
plt.rcParams.update({'font.size': 14}) # increase the font size
plt.xlabel('t')
plt.ylabel('N(t')
plt.title('Decaimento Radioativo')
plt.plot(t, N)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint


# Define a function which calculates the derivative
def dy_dx(y, x):
    return x - y

xs = np.linspace(-5,5,100)
y0 = 1.0 # teh initial condition
ys = odeint(dy_dx, y0, xs)
# ys = np.array(ys).flattern()

# Plot the numerical solution
plt.rcParams.update({'font.size': 14}) # increase the font size
plt.xlabel("x")
plt.ylabel('y')
plt.plot(xs, ys)
plt.show()
```

# EXEMPLO3 - presa/predador

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


# Modelo de presa-predador é também chamado de Lotka-Volterra (população de raposas e coelhos)
# Modelo de 2 variáveis e 4 parâmetros
# x -> presa e y -> predador
# a derivada da presa em relação a derivada do tempo é igual a ax - bxy
# a derivada do predador em relação a derivada do tempo é igual a -cy + e**xy

# Variável independente: t
# Variável dependente: x, y

# parâmetros
a = 0.7; b = 0.5; c = 0.3;  e = 0.2

#função da equação diferencial

def dP_dt(P, t):
    dx_dt = P[0]*(a - b*P[1]) # Presa
    dy_dt = -P[1]*(c - e*P[0]) # Predador
    return [dx_dt, dy_dt]

# vetor tempo
t = np.linspace(0, 20, 1000)

# condição inicial
P0 = [1.0, 0.5]

#aplicação do odeint
Ps = odeint(dP_dt, P0, t)

# separação das variáveis
prey = Ps[:,0]
predators = Ps[:,1]

#plot
#plt.style.use('classic')
plt.plot(t, prey,  label="Coelhos", linewidth = 2)
plt.plot(t, predators,  label="Raposas", linewidth = 2)
plt.xlabel("Tempo")
plt.ylabel("População")
plt.legend()
plt.show()
```

# Basic animation

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)

#anim.save('basic_animation.mp4', fps=30,writer='ffmpeg')

plt.show()
```
